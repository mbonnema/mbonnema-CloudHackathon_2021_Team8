```python
from harmony import BBox, Client, Collection, Request, LinkType
from harmony.config import Environment
import requests
from pprint import pprint
import datetime as dt
import s3fs
import xarray as xr
```

```python
url = 'https://cmr.earthdata.nasa.gov/search'
```

```python
collection_url = f'{url}/{"collections"}'
```

```python
short_name= 'HLSS30'
concept_id = 'C1711924822-LPCLOUD'
```

```python
response = requests.get(collection_url, 
                        params={
                            'short_name': short_name,
                            },
                        headers={
                            'Accept': 'application/json'
                            }
                       )
response = response.json()
pprint(response)
```

```python
granuleurl = f'{url}/{"granules"}'
response = requests.get(granuleurl, 
                        params={
                            'concept_id': concept_id,
                            'temporal': '2021-05-01T00:00:00Z,2021-05-10T23:59:59Z',
                            'bounding_box': '-114.941711,35.989119,-113.672791,36.582452',
                            'page_size': 200,
                            },
                        headers={
                            'Accept': 'application/json'
                            }
                       )
print(response.status_code)
print(response)
```

```python
print(response.headers['CMR-Hits'])
```

```python
granules = response.json()['feed']['entry']
```
